```python
# ðŸ“Š Explore Superstore Data â€” Final Step of ETL
import pandas as pd
import matplotlib.pyplot as plt

# Load the warehouse data (the final processed file)
df = pd.read_csv("../data/warehouse/sales_summary.csv")

# Quick overview
print("Rows:", len(df))
print("Columns:", df.columns.tolist())
df.head()
```

```python
region_sales = df.groupby("Region")["Sales"].sum().sort_values()
region_sales.plot(kind="barh", figsize=(8,4), title="Total Sales by Region")
plt.xlabel("Sales")
plt.show()
```

```python
df.columns
```

```python
region_sales = df.groupby("region")["total_sales"].sum().sort_values()
region_sales.plot(kind="barh", figsize=(8,4), title="Total Sales by Region", color="teal")
plt.xlabel("Total Sales")
plt.show()
```

```python
category_profit = df.groupby("category")["total_profit"].sum().sort_values(ascending=False)
category_profit.plot(kind="bar", color="orange", figsize=(6,4), title="Total Profit by Category")
plt.ylabel("Total Profit")
plt.show()
```

```python
avg_margin_region = df.groupby("region")["avg_profit_margin"].mean().sort_values(ascending=False)

avg_margin_region.plot(
    kind="bar",
    color="mediumseagreen",
    figsize=(6,4),
    title="Average Profit Margin by Region",
    edgecolor="black"
)
plt.ylabel("Average Profit Margin")
plt.show()
```

```python
### Insights Summary
- The **West region** has the highest total sales.
- The **Technology category** leads in profit.
- The **West region** shows the best average profit margin.
```
